```python
%matplotlib inline

import os

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from src.config import FIG_DIR, DATA_DIR
```

```python
def create_plot(data_filename, figure_filename, width, height, y_max, x_axis_title):

    plt.rc('pdf', fonttype=42)
    sns.set(style="ticks")
    sns.set_context('paper', font_scale=1.3)
    plt.subplots(figsize=(width, height))

    # my_colors = {'standard': 'lightgrey', 'content': 'darkgrey', 'hierarchical': 'grey', 'hybrid': 'dimgrey'}
    data = pd.read_csv(data_filename)
    # ax = sns.barplot(data=data, x="Metrics", y="Scores", palette=[my_colors[i] for i in data['Type']], alpha=0.9)
    ax = sns.barplot(data=data, x="Metrics", y="Scores", color='grey', alpha=0.9)
    ax.set_ylim(ymin=0, ymax=y_max)
    ax.grid(ls="dotted")
    ax.set_ylabel('Correlations', labelpad=15)
    ax.set_xlabel(x_axis_title, labelpad=15)
    sns.despine(offset=10, trim=True)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.savefig(figure_filename, dpi=200, transparent=False, bbox_inches='tight', pad_inches=0.0)
    plt.show()
```

```python
data_filename = os.path.join(FIG_DIR, 'us2_corr_pref.csv')
figure_filename = os.path.join(FIG_DIR, 'paper_us2_corr_pref.png')
create_plot(data_filename, figure_filename, 8, 4, 0.5, 'Evaluation metrics')
```

```python
data_filename = os.path.join(FIG_DIR, 'us2_corr_sim.csv')
figure_filename = os.path.join(FIG_DIR, 'paper_us2_corr_sim.png')
create_plot(data_filename, figure_filename, 4.5, 3.5, 0.5, 'Similarity functions')
```
